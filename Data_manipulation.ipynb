import libraries

```python
import pandas as pd
import numpy as np
```

CSV file

```python
student = {
    "Student": ["Riya", "Karan", "Meera", "Vivek", "Pooja", "Karan"],
    "Score": [88, 75, np.nan, 91, 84, 75],
    "Age": [19, 20, 21, np.nan, 20, 20]
}

df = pd.DataFrame(student)

df.to_csv("student_score.csv", index=False)

print("CSV File Created")
```

Data Load

```python
df = pd.read_csv("student_score.csv")

print(df)
```

Data Exploration

```python
print(df.head())

df.info()

print(df.describe())

print(df.isnull().sum())

print("Shape :", df.shape)

print("Size :", df.size)

print("Dimension :", df.ndim)
```

Handle missing values

```python
df["Score"] = df["Score"].fillna(df["Score"].mean())

df["Age"] = df["Age"].fillna(df["Age"].mean())
```

Remove duplicates

```python
print("Duplicate Rows :", df.duplicated().sum())

df.drop_duplicates(inplace=True)
```

Check Cleaned Data

```python
print(df)
```

Add Result Column

```python
df["Result"] = np.where(df["Score"] >= 80, "Pass", "Fail")

print(df)
```
