```python
# Install Libraries
!pip install -q sentence-transformers faiss-cpu

# Import Libraries
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Sample Knowledge Base
notes = [
    "Python is an easy and beginner-friendly programming language.",
    "Artificial Intelligence allows computers to solve intelligent problems.",
    "Machine Learning is a field of AI that learns from data.",
    "RAG combines document retrieval with language generation.",
    "Embeddings transform text into vector representations.",
    "FAISS performs fast similarity search using vector embeddings."
]

# Load Embedding Model
encoder = SentenceTransformer("all-MiniLM-L6-v2")

# Generate Embeddings
note_vectors = encoder.encode(notes)

# Create FAISS Index
vector_size = note_vectors.shape[1]

faiss_index = faiss.IndexFlatL2(vector_size)

faiss_index.add(np.array(note_vectors))

# User Question
question = input("Ask a Question: ")

# Convert Question into Vector
question_vector = encoder.encode([question])

# Search Top 2 Similar Documents
distances, indices = faiss_index.search(question_vector, 2)

print("\nRelevant Documents:\n")

retrieved_docs = []

for idx in indices[0]:
    print(notes[idx])
    retrieved_docs.append(notes[idx])

print("\nGenerated Answer:\n")

if retrieved_docs:
    answer = retrieved_docs[0]
    print(answer)
else:
    print("Sorry, no relevant information found.")
```
